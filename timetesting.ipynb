```python
def test1():
    for i in range(900):
        5**i
    print("Hello")
    for i in range(100):
        5**i
        
def test2():
    for i in range(1000):
        5**i
        if i == 100:
            print("Hello")
```

```python
%timeit test1()
```

```python
def test3():
    for i in range(1000):
        5**i

def test4():
    for i in list(range(1000)):
        5**i
```

```python
%timeit test3()
%timeit test4()
```

```python
balance = 0
total = 0

def command(a, b, c):
    global total
    global balance
    total += c
    balance += c
    
def recv_osc():
    global balance
    balance -= 1
    
events_in_buffer = 99
events_per_loop = 3001
total_events = 100010

total_events -= events_in_buffer

number_of_complete_loops = total_events // events_per_loop
number_of_events_in_last_loop = total_events % events_per_loop

command(31, 0, events_in_buffer)
for i in range(number_of_complete_loops):
    command(31, 0, events_per_loop)
    for j in range(events_per_loop):
        recv_osc()
command(31, 0, number_of_events_in_last_loop)
for j in range(events_in_buffer + number_of_events_in_last_loop):
    recv_osc()
        
print(total)
print(balance)
```
